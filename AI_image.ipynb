```python
!pip install diffusers accelerate torch gradio huggingface_hub

# ============================
# AI 이미지 생성기
# ============================

import torch
import gradio as gr
from diffusers import StableDiffusionPipeline

# 모델
def load_model():
    pipe = StableDiffusionPipeline.from_pretrained(
        "Lykon/AnyLoRA",
        torch_dtype=torch.float16
    )
    pipe = pipe.to("cuda")
    pipe.safety_checker = None
    pipe.requires_safety_checker = False
    return pipe

pipe = load_model()
print("모델 로딩 완료!")

# 이미지 생성
def generate(prompt, negative_prompt, steps):
    image = pipe(
        prompt,
        negative_prompt=negative_prompt,
        num_inference_steps=int(steps)
    ).images[0]
    return image

# Gradio
demo = gr.Interface(
    fn=generate,
    inputs=[
        gr.Textbox(
            label="프롬프트",
            value="1girl, full body, standing, white background, clean lineart, monochrome, anime style"
        ),
        gr.Textbox(
            label="네거티브 프롬프트",
            value="bad hands, extra fingers, bad feet, extra legs, bad anatomy, cropped, color, sketch, pencil, lowres, blurry"
        ),
        gr.Slider(10, 40, value=25, label="생성 단계"),
    ],
    outputs=gr.Image(label="결과"),
    title="AI 이미지 생성기",
    description="프롬프트를 입력하면 AI가 이미지를 생성합니다."
)

demo.launch()
```
